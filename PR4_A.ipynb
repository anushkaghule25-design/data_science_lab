```python
data = [1, 2, 4, 5, 6]

# --- Mean Calculation ---
sum_data = 0
count_data = 0
for x in data:
    sum_data += x
    count_data += 1

mean = sum_data / count_data
print(f"Dataset: {data}")
print(f"Mean: {mean}")

# --- Median Calculation ---
sorted_data = sorted(data)

if count_data % 2 == 1:
    median = sorted_data[count_data // 2]
else:
    mid1 = sorted_data[count_data // 2 - 1]
    mid2 = sorted_data[count_data // 2]
    median = (mid1 + mid2) / 2

print(f"Median: {median}")

# --- Variance Calculation ---
sum_squared_diff = 0
for x in data:
    diff = x - mean
    sum_squared_diff += diff * diff

# Using population variance (dividing by N)
variance = sum_squared_diff / count_data
print(f"Variance: {variance}")

standard_deviation = variance ** 0.5
print(f"Standard Deviation: {standard_deviation}")
```
