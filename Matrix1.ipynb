```python
# Create a 3x3 matrix using list
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

a = A[0][0]
b = A[0][1]
c = A[0][2]
d = A[1][0]
e = A[1][1]
f = A[1][2]
g = A[2][0]
h = A[2][1]
i = A[2][2]

det_A = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)
print("Matrix A:")
for row in A:
    print(row)
print("Determinant of A:", det_A)
```
