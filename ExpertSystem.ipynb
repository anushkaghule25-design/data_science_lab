```python
# Simple Disease Expert System

print("Simple Disease Diagnosis System")

fever = input("Do you have fever? (yes/no): ")
cough = input("Do you have cough? (yes/no): ")
headache = input("Do you have headache? (yes/no): ")

if fever == "yes" and cough == "yes":
    print("Diagnosis: You may have Flu.")
elif fever == "yes" and headache == "yes":
    print("Diagnosis: You may have Viral Fever.")
elif cough == "yes" and headache == "yes":
    print("Diagnosis: You may have Common Cold.")
else:
    print("Diagnosis: Symptoms unclear. Please consult a doctor.")
```
