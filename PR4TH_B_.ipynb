```python
import numpy as np
import matplotlib.pyplot as plt

mean = 60
std_dev = 20
size = 1000

data = np.random.normal(mean, std_dev, size)

plt.hist(data, bins=30)
plt.title("Histogram of Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
