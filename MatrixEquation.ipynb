```python
import numpy as np

# Coefficient matrix
A = np.array([
    [2, 3],
    [4, 5]
])

# Constant values
B = np.array([10, 20])

# Solve the equations
solution = np.linalg.solve(A, B)

print("Value of x =", solution[0])
print("Value of y =", solution[1])
```
