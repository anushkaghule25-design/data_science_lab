```python

def a_star(graph, heuristic, start, goal):
    open_list = [start]          # Nodes to be evaluated
    closed_list = []             # Nodes already evaluated
    g = {start: 0}               # Cost from start to node
    parent = {start: None}       # Store path
    while open_list:
        # Choose node with lowest f = g + h
        current = min(open_list, key=lambda node: g[node] + heuristic[node])
        # Goal reached
        if current == goal:
            path = []
            while current is not None:
                path.append(current)
                current = parent[current]
            path.reverse()
            return path
        open_list.remove(current)
        closed_list.append(current)
        for neighbor, cost in graph[current]:
            if neighbor in closed_list:
                continue
            new_cost = g[current] + cost
            if neighbor not in open_list or new_cost < g.get(neighbor, float('inf')):
                g[neighbor] = new_cost
                parent[neighbor] = current
                if neighbor not in open_list:
                    open_list.append(neighbor)
    return "No path found"
graph = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 1)],
    'C': [('D', 1)],
    'D': []
}

heuristic = {
    'A': 3,
    'B': 2,
    'C': 1,
    'D': 0
}
start = 'A'
goal = 'D'
result = a_star(graph, heuristic, start, goal)
print("Shortest Path:", result)
```
