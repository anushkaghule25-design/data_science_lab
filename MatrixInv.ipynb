```python
# Create 3x3 matrix using list
B = [
    [2, 1, 1],
    [1, 3, 2],
    [1, 0, 0]
]

# Assign elements
a = B[0][0]
b = B[0][1]
c = B[0][2]
d = B[1][0]
e = B[1][1]
f = B[1][2]
g = B[2][0]
h = B[2][1]
i = B[2][2]

# Step 1: Find Determinant
det = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)

print("Matrix B:")
for row in B:
    print(row)

print("Determinant of B:", det)

if det == 0:
    print("Inverse does not exist")
else:
    # Step 2: Find Cofactors
    A11 = (e*i - f*h)
    A12 = -(d*i - f*g)
    A13 = (d*h - e*g)

    A21 = -(b*i - c*h)
    A22 = (a*i - c*g)
    A23 = -(a*h - b*g)

    A31 = (b*f - c*e)
    A32 = -(a*f - c*d)
    A33 = (a*e - b*d)

    # Step 3: Adjoint (transpose of cofactor matrix)
    adj = [
        [A11, A21, A31],
        [A12, A22, A32],
        [A13, A23, A33]
    ]


    inverse = []
    for row in adj:
        inverse_row = []
        for element in row:
            inverse_row.append(element / det)
        inverse.append(inverse_row)

    print("Inverse of B:")
    for row in inverse:
        print(row)
```
