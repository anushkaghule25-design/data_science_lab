```python
import numpy as np

A = np.array([[2, 3],
              [4, 5]])

B = np.array([10, 20])

solution = np.linalg.solve(A, B)

print("Given Equations:")
print("2x + 3y = 10")
print("4x + 5y = 20")

print("\nSolution:")
print("x =", solution[0])
print("y =", solution[1])
```
