```python
 import pandas as pd
 import numpy as np
```

```python
df=pd.read_csv("/BostonHousing.csv")
print(df.head())
```

```python
'''Feature	Description
CRIM	Per capita crime rate by town
ZN	Proportion of residential land zoned
INDUS	Proportion of non-retail business acres
CHAS	Charles River dummy variable (1 if near river, 0 otherwise)
NOX	Nitric oxide concentration
RM	Average number of rooms per dwelling
AGE	Proportion of owner-occupied units built before 1940
DIS	Distance to employment centers
TAX	Property tax rate
PTRATIO	Pupil-teacher ratio
LSTAT	% lower status population
MEDV	Median value of homes (in $1000s)'''
```

```python
print("Shape of dataset:", df.shape)
```

```python
df.isnull().sum()
```

```python
print("\nColumn Names:\n", df.columns)
```

```python
print("\nSummary Statistics:\n", df.describe())
```

```python
houses_more_than_6_rooms = df[df['rm'] > 6]
print("Houses with more than 6 rooms:")
print(houses_more_than_6_rooms.head())
```

```python
expensive_houses = df[df['medv'] > 25]
print("Expensive Houses (MEDV > 25):")
print(expensive_houses.head())

```

```python
river_houses = df[df['chas'] == 1]
print("Houses near Charles River:")
print(river_houses.head())
```

```python
print("Mean MEDV:", df['medv'].mean())
print("Minimum MEDV:", df['medv'].min())
print("Maximum MEDV:", df['medv'].max())
print("Sum of MEDV:", df['medv'].sum())
print("Count of MEDV:", df['medv'].count())
```

```python
print("Mean AGE:", df['age'].mean())
print("Minimum AGE:", df['age'].min())
print("Maximum AGE:", df['age'].max())
print("Sum of AGE:", df['age'].sum())
print("Count of AGE:", df['age'].count())
```

```python
group_chas = df.groupby('chas')['medv'].agg(['count', 'mean', 'min', 'max', 'sum'])
print(group_chas)
```
