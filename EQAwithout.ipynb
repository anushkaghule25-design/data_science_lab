```python
print("Given Equations:")
print("2x + 3y = 10")
print("4x + 5y = 20")

a1, b1, c1 = 2, 3, 10
a2, b2, c2 = 4, 5, 20

y = (c1*a2 - c2*a1) / (b1*a2 - b2*a1)
x = (c1 - b1*y) / a1

print("\nSolution:")
print("x =", x)
print("y =", y)
```
