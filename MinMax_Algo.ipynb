```python
# Minimax function
def minimax(depth, nodeIndex, isMax, scores, maxDepth):

    # Base case: if we reach leaf node
    if depth == maxDepth:
        return scores[nodeIndex]

    # If this is MAX player's turn
    if isMax:
        return max(
            minimax(depth + 1, nodeIndex * 2, False, scores, maxDepth),
            minimax(depth + 1, nodeIndex * 2 + 1, False, scores, maxDepth)
        )

    # If this is MIN player's turn
    else:
        return min(
            minimax(depth + 1, nodeIndex * 2, True, scores, maxDepth),
            minimax(depth + 1, nodeIndex * 2 + 1, True, scores, maxDepth)
        )


# Driver code
scores = [3, 5, 2, 9]   # Leaf node values
maxDepth = 2           # Depth of the tree

bestValue = minimax(0, 0, True, scores, maxDepth)

print("The optimal value is:", bestValue)
```
