```python
# Coefficients
a1 = 2
b1 = 3
c1 = 10

a2 = 4
b2 = 5
c2 = 20

a1_new = a1 * 2
b1_new = b1 * 2
c1_new = c1 * 2

# Subtract second equation
y = (c1_new - c2) / (b1_new - b2)

# Substitute y in first equation
x = (c1 - b1 * y) / a1

print("Value of x =", x)
print("Value of y =", y)
```
