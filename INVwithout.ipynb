```python
def inverse_3x3(m):
    a, b, c = m[0]
    d, e, f = m[1]
    g, h, i = m[2]

    det = (a*(e*i - f*h)
           - b*(d*i - f*g)
           + c*(d*h - e*g))

    if det == 0:
        return None

    inv = [
        [(e*i - f*h)/det, (c*h - b*i)/det, (b*f - c*e)/det],
        [(f*g - d*i)/det, (a*i - c*g)/det, (c*d - a*f)/det],
        [(d*h - e*g)/det, (b*g - a*h)/det, (a*e - b*d)/det]]

    return inv
matrix = [
    [2, 1, 1],
    [1, 3, 2],
    [1, 0, 0]
]

result = inverse_3x3(matrix)

if result is None:
    print("Matrix is not invertible.")
else:
    print("Inverse matrix:")
    for row in result:
        print(row)
```
