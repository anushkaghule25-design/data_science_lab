```python
import pandas as pd
import numpy as np
# Load dataset (Boston Housing Dataset from Kaggle)
df = pd.read_csv("/content/Boston Housing.csv")
# Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())
# Houses with higher room values
filtered_rooms = df[df['RM'] > 0.5]
print("\nHouses with higher room values:")
print(filtered_rooms.head())
# Houses with higher price values
filtered_price = df[df['MEDV'] > 3]
print("\nHouses with higher price values:")
print(filtered_price.head())
# ---------------- Grouping & Aggregation ----------------
grouped_data = df.groupby('CHAS')['MEDV'].agg(['count','sum','mean','min','max'])

print("\nAggregate statistics of MEDV grouped by CHAS:")
print(grouped_data)
```
