```python
import numpy as np

# Create a 3x3 matrix
B = np.array([
    [2, 1, 1],
    [1, 3, 2],
    [1, 0, 0]
])

# Find inverse
B_inverse = np.linalg.inv(B)

print("Matrix B:")
print(B)
print("Inverse of B:")
print(B_inverse)
```
