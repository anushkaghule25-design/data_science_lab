```python
# Graph example
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# DFS function
def dfs(graph, node, visited=None):
    if visited is None:
        visited = []

    if node not in visited:
        visited.append(node)

        for neighbor in graph[node]:
            dfs(graph, neighbor, visited)

    return visited

# Run DFS
result = dfs(graph, 'A')
print("DFS Traversal:", result)
```
