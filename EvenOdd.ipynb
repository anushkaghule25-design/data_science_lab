```python
def check_odd_or_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 10
print(check_odd_or_even(num1))

num2 = 7
print(check_odd_or_even(num2))

num3 = 0
print(check_odd_or_even(num3))
```
