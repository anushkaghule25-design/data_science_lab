```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("The 3x3 matrix is:")
for row in matrix:
    print(row)

a = matrix[0][0]
b = matrix[0][1]
c = matrix[0][2]
d = matrix[1][0]
e = matrix[1][1]
f = matrix[1][2]
g = matrix[2][0]
h = matrix[2][1]
i = matrix[2][2]

determinant = a * (e * i - f * h) - b * (d * i - f * g) + c * (d * h - e * g)

print(f"\nThe determinant of the matrix is: {determinant}")
```
